```python
# creating tensors with different datatype
float16_tensor = torch.tensor([3.0, 4.0,5.6], dtype=torch.half)
float16_tensor.dtype
```

```python
# basic operations on tensors
tensor = torch.tensor([1,2,3])
tensor + 10, tensor*10, tensor-10
# we can also use torch.mul, torch.add
# tensors valuse do not change until they have been resassigned
```

```python
a = torch.tensor([1,2,3])
b = torch.tensor([[3],[4],[5]])
torch.matmul(a,b)
# can also use '@' but not recommended
# matmul is faster compared to doing this via for loop
# can also use torch.mm which is short of matmul
```

```python
# aggregation function
demo_tensor = torch.arange(0,100,10)
demo_tensor.min(), demo_tensor.max(), demo_tensor.sum(), demo_tensor.type(torch.float32).mean()
# mean() do not work without datatype and throws an error
#we can also use torch.max(demo_tensor)
#positional min and max
torch.argmin(demo_tensor)
```

```python
# change tensor datatype
tensor = torch.tensor([21.4, 22.5, 56.7])
tensor = tensor.type(torch.float16)
tensor = tensor.type(torch.int8)
tensor.dtype, tensor
```
